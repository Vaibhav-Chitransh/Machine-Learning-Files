```python
ds1 = pd.read_csv('iris.csv')
ds1

ds1 = ds1.fillna(ds1.mean())

# is data me label encoding pehle se hi kar rakhi hai
```

```python
x = ds1.iloc[:, :-1].values
y = ds1.iloc[:, -1].values
print(x, y)
```

```python
# Code for testing and training the data
from sklearn.model_selection import train_test_split

# Split data into train and test sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)
```

```python
# Code for precprocessing the dataset
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier()
classifier.fit(x_train, y_train)
```

```python
y_pred = classifier.predict(x_test)
print(np.concatenate((y_pred.reshape(len(y_pred), 1), y_test.reshape(len(y_test), 1)), 1))
```

```python
new_data_point = [[0.0, 4, 3, 0.4]]
transformed_data_point = sc.transform(new_data_point)
predictions = classifier.predict(transformed_data_point)
print(predictions)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```
