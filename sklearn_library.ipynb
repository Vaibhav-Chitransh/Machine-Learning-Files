```python
data_1 = {'Name':['Vaibhav' , 'Rishi', 'Gagan', 'Anshu'],
          'Age' : [20, 21, 21, np.nan],
          'Occupation': ['MERN Stack Developer', 'Programmer', 'Android Developer', 'UI/UX Designer'],
          'HomeTown': ['Jaipur', 'Chennai', 'Varanasi', 'Gwalior'],
          'NetWorth' : ['120 billion', '100 billion', '390 million', '80 billion']}

dataset1 = pd.DataFrame(data_1)
dataset1
```

```python
occupation_col = dataset1.loc[0:3, 'Occupation']
print(occupation_col)
```
